```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
drug_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merged_data = mouse_data.merge(drug_data)

# Display the data table for preview
merged_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_tumor_volume = merged_data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
mean_tumor_volume = pd.DataFrame(grouped_tumor_volume).reset_index()

# Preview DataFrame
mean_tumor_volume.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_volume_SE = merged_data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].sem()

# Convert to DataFrame
tumor_volume_SE_df = pd.DataFrame(tumor_volume_SE).reset_index()

# Preview DataFrame
tumor_volume_SE_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tumor_volume_pivot = mean_tumor_volume.pivot\
(
    index='Timepoint',
    columns='Drug',
    values='Tumor Volume (mm3)'
)

tumor_volume_SE_pivot = tumor_volume_SE_df.pivot\
(
    index='Timepoint',
    columns='Drug',
    values='Tumor Volume (mm3)'
)

# Preview that Reformatting worked
print(mean_tumor_volume_pivot.head())
print(tumor_volume_SE_pivot.head())
```

```python
# Generate the Plot (with Error Bars)

markerType = ['1','x',4,'o','s','*','h','d','P','|']   #Select data markers

fig, ax = plt.subplots(figsize=(7,9))   #Set figure size/dimensions to best show data

for i in range(len(mean_tumor_volume_pivot.columns)):   #Generate a line graph for each set of data
    ax.errorbar\
    (
        mean_tumor_volume_pivot.index,   #Define x values
        mean_tumor_volume_pivot[mean_tumor_volume_pivot.columns[i]],   #Define y values
        tumor_volume_SE_pivot[tumor_volume_SE_pivot.columns[i]],   #Define y error values
        label = mean_tumor_volume_pivot.columns[i],   #Label each data set
        markersize=7,   #Set marker size
        marker=markerType[i]  #Set marker type from list above
    )

ax.set_xlim(0,46)

ax.set_xlabel('Time (Days)')
ax.set_ylabel('Tumor Volume (mm3)')

ax.set_title('Tumor Resonse to Treatment')

ax.legend(loc=2, fontsize='medium', ncol=2)

ax.grid(axis='y')

plt.show()
# Save the Figure

fig.savefig('tumor_response.png')
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met = merged_data.groupby(['Drug','Timepoint'])['Metastatic Sites'].mean()

# Convert to DataFrame
mean_met_df = pd.DataFrame(mean_met).reset_index()

# Preview DataFrame
mean_met_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
mean_met_SE = merged_data.groupby(['Drug','Timepoint'])['Metastatic Sites'].sem()

# Convert to DataFrame
mean_met_SE_df = pd.DataFrame(mean_met_SE).reset_index()

# Preview DataFrame
mean_met_SE_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_met_pivot = mean_met_df.pivot\
(
    index='Timepoint',
    columns='Drug',
    values='Metastatic Sites'
)

mean_met_SE_pivot = mean_met_SE_df.pivot\
(
    index='Timepoint',
    columns='Drug',
    values='Metastatic Sites'
)

# Preview that Reformatting worked
print(mean_met_pivot.head())
print(mean_met_SE_pivot.head())
```

```python
# Generate the Plot (with Error Bars)
# Comments same as in last figure
markerType = ['1','x',4,'o','s','*','h','d','P','|']

fig, ax = plt.subplots(figsize=(7,9))

for i in range(len(mean_met_pivot.columns)):
    ax.errorbar\
    (
        mean_met_pivot.index, 
        mean_met_pivot[mean_met_pivot.columns[i]], 
        mean_met_SE_pivot[mean_met_SE_pivot.columns[i]],
        label = mean_met_pivot.columns[i],
        markersize=7,
        marker=markerType[i]
    )

ax.set_xlim(0,46)

ax.set_xlabel('Time (Days)')
ax.set_ylabel('Metastatic Sites')

ax.set_title('Metastatic Spread During Treatment Treatment')

ax.legend(loc=2, fontsize='medium', ncol=2)

ax.grid(axis='y')

# Save the Figure
fig.savefig('metastatic_spread.png')

# Show the Figure

fig.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count = merged_data.groupby(['Drug','Timepoint'])['Mouse ID'].count()

# Convert to DataFrame
mouse_count_df = pd.DataFrame(mouse_count).reset_index()
mouse_count_df = mouse_count_df.rename(columns={'Mouse ID':'Mouse Count'})

# Preview DataFrame
mouse_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_pivot = mouse_count_df.pivot\
(
    index='Timepoint',
    columns='Drug',
    values='Mouse Count'
)

# Preview the Data Frame
mouse_count_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
# Comments the same as in previous figures
markerType = ['1','x',4,'o','s','*','h','d','P','|']

fig, ax = plt.subplots(figsize=(7,9))

for i in range(len(mouse_count_pivot.columns)):
    ax.plot\
    (
        mouse_count_pivot.index, 
        mouse_count_pivot[mouse_count_pivot.columns[i]] / mouse_count_pivot[mouse_count_pivot.columns[i]][0] * 100,
        label = mouse_count_pivot.columns[i],
        markersize=7,
        marker=markerType[i]
    )

ax.set_xlim(0,46)

ax.set_xlabel('Time (Days)')
ax.set_ylabel('Survival Rate (%)')

ax.set_title('Survival During Treatment')

ax.legend(loc=3, fontsize='medium', ncol=2)

ax.grid()

plt.show()

# Save the Figure
fig.savefig('survival.png')

# Show the Figure
fig.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
pct_chg = \
[
    -100 + 
    mean_tumor_volume_pivot[mean_tumor_volume_pivot.columns[i]].values[-1] / 
    mean_tumor_volume_pivot[mean_tumor_volume_pivot.columns[i]].values[0] * 
    100 for i in range(len(mean_tumor_volume_pivot))
]

# Display the data to confirm
pct_chg
```

```python
# Calculate the propagation of error for each value using equation 11 from https://chem.libretexts.org/Bookshelves/Analytical_Chemistry/Supplemental_Modules_(Analytical_Chemistry)/Quantifying_Nature/Significant_Digits/Propagation_of_Error#Arithmetic_Error_Propagation

# Relative error cllulated as the root of the sum of the squares of the ratio of SE to value for each point
pct_chg_relative_error = \
[
    (
        (tumor_volume_SE_pivot[mean_tumor_volume_pivot.columns[i]].values[0]/mean_tumor_volume_pivot[mean_tumor_volume_pivot.columns[i]].values[0])**2 + 
        (tumor_volume_SE_pivot[mean_tumor_volume_pivot.columns[i]].values[-1]/mean_tumor_volume_pivot[mean_tumor_volume_pivot.columns[i]].values[-1])**2
    )**(0.5) 
    for i in range(len(pct_chg))
]

# Actual error calulated using the percent error above
pct_chg_error = \
[
    pct_chg[i] * 
    pct_chg_relative_error[i] 
    for i in range(len(pct_chg))
]

# Define arrays containg error values for negative and positive change
neg_pct_chg_error = []
pos_pct_chg_error = []

# Sort values and populate arrays
for i in sorted(pct_chg_error):
    if i > 0:
        pos_pct_chg_error.append(i)
    elif i  < 0:
        neg_pct_chg_error.append(i)
```

```python
# Store all Relevant Percent Changes into a Tuple
pct_chg_df = pd.DataFrame({'Percent Change':pct_chg}, index=mean_tumor_volume_pivot.columns)   # Create data frame from data
pct_chg_df = pct_chg_df.sort_values(by='Percent Change').reset_index()   # Sort data ascending by percent change and reset indexes

# Splice the data between passing and failing drugs
fig, ax = plt.subplots(figsize=(12,6))

# Bars for drugs with negative change
neg_chg = ax.bar\
(
    pct_chg_df[pct_chg_df['Percent Change'] < 0]['Drug'],
    pct_chg_df[pct_chg_df['Percent Change'] < 0]['Percent Change'],
    align='edge',
    width=0.97,
    color='green',
    yerr=neg_pct_chg_error
)

# Bars for drugs with positive change
pos_chg = ax.bar\
(
    pct_chg_df[pct_chg_df['Percent Change'] > 0]['Drug'],
    pct_chg_df[pct_chg_df['Percent Change'] > 0]['Percent Change'],
    align='edge',
    width=0.97,
    color='red',
    yerr=pos_pct_chg_error
)

# Orient widths. Add labels, tick marks, etc. 
ax.set_xlim(-0.05,10)
ax.set_ylabel('% Tumor Volume Change', fontsize='large')
ax.set_title('Tumor Change Over 45 Day Treatment', fontsize='x-large', weight='bold')

plt.xticks(ha='left')
plt.grid()

# Use functions to label the percentages of changes
def posChgLabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate\
        (
            '{:.1f}%'.format(height),
            xy=(rect.get_x() + rect.get_width() / 2, 0),
            xytext=(0, 5), 
            textcoords="offset points",
            ha='center', va='bottom',
            color='white', weight='bold',
            path_effects=[path_effects.withStroke(linewidth=2, foreground='black')]   # Use path_effects module to outline labels to make them more visible
        )

def negChgLabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate\
        (
            '{:.1f}%'.format(height),
            xy=(rect.get_x() + rect.get_width() / 2, 0),
            xytext=(0, -5),  # 3 points vertical offset
            textcoords="offset points",
            ha='center', va='top',
            color='white', weight='bold',
            path_effects=[path_effects.withStroke(linewidth=2, foreground='black')]   # Use path_effects module to outline labels to make them more visible
        )

# Call functions to implement the function calls
negChgLabel(neg_chg)
posChgLabel(pos_chg)

# Save the Figure
fig.savefig('tumor_change.png')

# Show the Figure
fig.show()
```

# Three Observations from Data

* Depending on what type of graph is used, it is not immediately obvious which drug(s) is(are) most effective.  For example while there is no obvious difference between drugs when viewing the graph for Metastatic Spread During Treatment, it is very clear which drugs are most beneficial when viewing the bar graph of Tumor Change Over 45 Day Treatment.  This could indicate that increases in metastatic sites do not correlate strongly with drug effectivness.

* The largest errors and least percision can be seen in data with a small range and accuracy (whether due to measurment or experiment design).  For example when comparing changes in tumor volume, each data point had an accuracy to 8 significant digits and a spread in data both above and below the initial starting point, whereas the metastatic site spread during treatment or survival rate data had an accuracy to 1 siginificant digit.  In situations like this a larger sample size is essential in order to get meaningful information from the data.

* There was no "middle ground" for effectiveness of a drug, all drugs were fairly clustered when comparing survival rate or tumor volume change and were either effective or not.
